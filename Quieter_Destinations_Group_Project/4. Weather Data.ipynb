## Obtaining historic weather data using OpenWeather API ##

```python
import requests
import pandas as pd
import numpy as np
```

```python
# Using OpenWeather API to obtain weather data
base_url = "https://api.openweathermap.org/data/3.0/onecall/day_summary" 
api_key = "*****************************" # Enter your API key

# Function that requests and returns data from the OpenWeather API
def data(lat, lon, date):
    url = f"{base_url}?lat={lat}&lon={lon}&date={date}&appid={api_key}&units=metric"
    response = requests.get(url)

    if response.status_code == 200:
        requested_data = response.json()
        return requested_data
    else:
        print("Not found")
```

```python
# Trialling weather function to see if it is successful in skipping index 2 (American Samoa)
dates = ['2024-06-06', '2023-06-06', '2022-06-06', '2021-06-06']

def trial_index_avg_weather(dates):
    temps = [] 
    total_rain = []

    # Ensuring index 2 gets looped through
    for i in range(3):
        lat = df.loc[i,'latitude']
        lon = df.loc[i,'longitude']

        # index 2 should be skipped in if statement
        if i in nogo_list:
            pass 
        else:
            for j in dates:
                date = j
                weather_data = data(lat, lon, date)
                temp_12pm = (weather_data['temperature']['afternoon']) 
                rain = (weather_data['precipitation']['total'])

                temps.append(temp_12pm)
                total_rain.append(rain)
            
                avg_temp = round(sum(temps) / len(temps))        
                avg_rainfall = round(sum(total_rain) / len(total_rain), 1)
        
            df.loc[i, 'avg_temp'] = avg_temp
            df.loc[i, 'avg_rainfall'] = avg_rainfall


trial_index_avg_weather(dates)
```

Need to get weather data for Greece separately as weather data for new, quieter destinations need to be compared to Greece 

```python
# Finding index in dataframe for Greece
gr_index = df.index[df['Country'] == 'Greece'].tolist()

# Creating a list for the index so this can be iterated through in weather function
old_dest = []
for i in gr_index:
    old_dest.append(i)

print(old_dest)
```

```python
# Using function to get weather data for Greece
dates = ['2024-06-06', '2023-06-06', '2022-06-06', '2021-06-06']

def gr_weather(dates):
    temps = [] 
    total_rain = []

    for i in old_dest:
        lat = df.loc[i,'latitude']
        lon = df.loc[i,'longitude']
        for j in dates:
            date = j
            weather_data = data(lat, lon, date)
            temp_12pm = (weather_data['temperature']['afternoon']) 
            rain = (weather_data['precipitation']['total'])

            temps.append(temp_12pm)
            total_rain.append(rain)

            # Calculating mean averages of temperatures and rainfall
            avg_temp = round(sum(temps) / len(temps))        
            avg_rainfall = round(sum(total_rain) / len(total_rain), 1)
        
        df.loc[i, 'avg_temp'] = avg_temp
        df.loc[i, 'avg_rainfall'] = avg_rainfall


gr_weather(dates)
```
