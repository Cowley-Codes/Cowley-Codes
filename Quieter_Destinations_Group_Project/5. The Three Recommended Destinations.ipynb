# Finding the top three destinations for our solo female traveller #

Using the Completed_Weather_Data_For_Destinations.csv file, filtered out the top destinations for our traveller by the weather (+- 2 degrees compared to Greece's average temp, and either equal to or lower than Greece's average rainfall)

```python
import pandas as pd
```

```python
df = pd.read_csv('project_data/Complete_Weather_Data_For_Destinations.csv')

df.head()
```

```python
greece_df = df[df['Country'] == 'Greece']
```

```python
greece_df.head()
```

```python
greece_data = df[df['Country'] == 'Greece'].iloc[0]
```

```python
greece_temp = greece_data['Avg Temp(째C)']
greece_rain = greece_data['Avg Rainfall(mm)']
```

```python
# Filtering the countries based on temperature and rainfall in comparison to Greece.
top_destinations_df = df[
    (df['Avg Temp(째C)'] >= greece_temp - 2) &
    (df['Avg Temp(째C)'] <= greece_temp + 2) &
    (df['Avg Rainfall(mm)'] <= greece_rain)
]
```

```python
top_destinations_df.head()
```

```python
# Ordering the 5 countries from top_destinations_df from lowest total arrivals to highest

ordered_top_destinations = top_destinations_df.sort_values('Mean Total Arrivals').reset_index(drop=True)
```

```python
ordered_top_destinations.head()
```

## The top three destinations are Aruba, Azerbaijan and Australia ##

This is based on finding destinations comparable to the climate of Greece (+- 2 degrees compared to Greece's average temperature, and either equal to or lower than Greece's average rainfall), whilst finding the quietest destinations based on mean total arrivals.
